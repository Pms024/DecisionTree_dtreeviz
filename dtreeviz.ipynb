```python
from sklearn.datasets import *
from sklearn import tree
import dtreeviz 
import numpy as np
```

### Classification

```python
# Sample iris data form sklearn
clas = tree.DecisionTreeClassifier()
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

```python
viz_c = dtreeviz.model(clas,
               X_train,
               y_train,
               feature_names = iris.feature_names,
               class_names = ['setosa', 'versicolor', 'virginica'])
```

```python
viz_c.view(scale=1.5)
```

```python
# Selecting particular row values
X = iris.data[np.random.randint(len(iris.data)),:]
print(iris.feature_names)
print(X)
```

```python
viz_c.view(x=X,scale=1.3) # also visualize only predicted path using - show_just_path='True'
```

```python
# To change orientation left-right
viz_c.view(orientation='LR') #default 'TD'
```

```python
# Feature importance for X instance
viz_c.instance_feature_importance(x=X)
```

### Regression

```python
reg = tree.DecisionTreeRegressor(max_depth=2)
boston = load_boston()
```

```python
X_train = boston.data
y_train = boston.target
```

```python
reg.fit(X_train,y_train)
```

```python
viz_r = dtreeviz.model(reg,
                       X_train,
                       y_train,
                       feature_names=boston.feature_names,
                       target_name='price')
```

```python
viz_r.view(scale=2)
```

```python
#Selecting random row
X = boston.data[np.random.randint(0,len(boston.data)),:]
X
```

```python
viz_r= dtreeviz.model(reg,
                      X_train,
                      y_train,
                      feature_names=boston.feature_names,
                      target_name='price')
```

```python
viz_r.view(x=X,scale=2) # Visualizing the prediction path of X instance
```

```python
# string representation of X prediction path
viz_r.explain_prediction_path(x=X)
```

```python
# leaf purity 
viz_r.leaf_purity()
```
